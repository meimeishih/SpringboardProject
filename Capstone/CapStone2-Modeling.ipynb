```python
from sklearn.model_selection import train_test_split
TX_train, TX_test,Ty_train, Ty_test=train_test_split(TraOx,TraOy,test_size=0.3)

HX_train, HX_test,Hy_train, Hy_test=train_test_split(HOx,HOy,test_size=0.3)

FX_train, FX_test,Fy_train, Fy_test=train_test_split(FOx,FOy,test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn.model_selection import validation_curve

scaler=StandardScaler()
scaler.fit(TX_train)
TX_train=scaler.transform(TX_train)
TX_test=scaler.transform(TX_test)

scaler=StandardScaler()
scaler.fit(HX_train)
HX_train=scaler.transform(HX_train)
HX_test=scaler.transform(HX_test)

scaler=StandardScaler()
scaler.fit(FX_train)
FX_train=scaler.transform(FX_train)
FX_test=scaler.transform(FX_test)

```

```python
Ty_pred_proba = TLR.predict_proba(TX_test)[::,1]
Tauc = metrics.roc_auc_score(Ty_test, Ty_pred_proba)
print(Tauc)
```

```python
# Historical Data Only
HLR=LogisticRegression(max_iter=5000)
HLR.fit(HX_train,Hy_train)
print(accuracy_score(HLR.predict(HX_test),Hy_test))
```

```python
#AUC for Historical Data Only
Hy_pred_proba = HLR.predict_proba(HX_test)[::,1]
Hauc = metrics.roc_auc_score(Hy_test, Hy_pred_proba)
print(Hauc)
```

```python
#Full Dataset
FLR=LogisticRegression(max_iter=5000)
FLR.fit(FX_train,Fy_train)
print(accuracy_score(FLR.predict(FX_test),Fy_test))
```

```python
# AUC for Full Dataset
Fy_pred_proba = FLR.predict_proba(FX_test)[::,1]
Fauc = metrics.roc_auc_score(Fy_test, Fy_pred_proba)
print(Fauc)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc

#Gradient Boosting for Triage Only - Find best Learning Rate
learning_rates = [0.05, 0.1, 0.25, 0.5, 0.75, 1]
for learning_rate in learning_rates:
    Tgb = GradientBoostingClassifier(learning_rate = learning_rate)
    Tgb.fit(TX_train, Ty_train)
    print("Learning rate: ", learning_rate)
    print("Accuracy score (training): {0:.3f}".format(Tgb.score(TX_train, Ty_train)))
    print("Accuracy score (validation): {0:.3f}".format(Tgb.score(TX_test, Ty_test)))
    print()
```
