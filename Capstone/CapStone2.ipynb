```python
ESIDF=df1[['esi','disposition']]
print(ESIDF)
```

```python
ESIDF.groupby('disposition')['esi'].value_counts().unstack().plot.bar(stacked=False)
```

From the graph above, we can see that Majority of patient with esi 4 and 5 were direct discharge from ED. People with esi 1 and 2 admit to unit

```python
df1.groupby('disposition')['triage_vital_o2_device'].value_counts().unstack().plot.bar(stacked=False)
```

```python
df1.boxplot(by='disposition',column=['triage_vital_hr','triage_vital_sbp','triage_vital_dbp','triage_vital_rr','triage_vital_o2','triage_vital_temp'],figsize=(30,30),grid=False)
```

The median of triage vitals show not much difference from Admit to Discharge. However, the range of the vital is larger in Admit patients.
